```python
# Extracting classes
buys_computer = ['no', 'yes']
# print(buys_computer)
# Count occurrences of each class
class_counts = {}
total_instances = len(buys_computer)

for value in buys_computer:
    if value in class_counts:
        class_counts[value] += 1
    else:
        class_counts[value] = 1

# Calculate prior probabilities
prior_probabilities = {key: value / total_instances for key, value in class_counts.items()}

print("Prior probabilities for each class:")
print(prior_probabilities)
```

```python
from collections import defaultdict


# Extracting features
features = [row[:-1] for row in data]
print(features)
# Group data by class
grouped_data = defaultdict(list)
for i, value in enumerate(buys_computer):
    grouped_data[value].append(features[i])

# Calculate class conditional densities
class_conditional_densities = {}
for class_value, instances in grouped_data.items():
    class_data = list(zip(*instances))
    class_conditional_densities[class_value] = []

    for feature_values in class_data:
        feature_density = {}

        for value in set(feature_values):
            feature_density[value] = feature_values.count(value) / len(feature_values)
        
        class_conditional_densities[class_value].append(feature_density)

print("Class conditional densities:")
print(class_conditional_densities)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split

features = data.drop(columns=['age', 'buys_computer'], axis=1)
buys_computer = data['buys_computer']
```

```python
features
```

```python
# replace values with numbers

features = features.replace({'high': 0, 'medium': 1, 'low': 2})
features = features.replace({'excellent': 0, 'fair': 1})
features = features.replace({'yes': 0, 'no': 1})

features
```

```python


# label encoder buys_computer

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
buys_computer = le.fit_transform(buys_computer)
# features = le.fit_transform(features)

#
# Split the data into training and testing sets
Tr_X, Te_X, Tr_y, Te_y = train_test_split(features, buys_computer, test_size=0.3, random_state=42)

# Assuming Tr_X, Tr_y, Te_X, Te_y are your training and testing data
# Replace these placeholders with your actual data

# Instantiate the Gaussian Naïve Bayes model
model = GaussianNB()

# Train the model
model.fit(Tr_X, Tr_y)

# Predict on the test set
predictions = model.predict(Te_X)

# Calculate accuracy
accuracy = accuracy_score(Te_y, predictions)
print(f"Accuracy of the Naïve Bayes classifier: {accuracy}")
print(classification_report(Te_y, predictions))

```
