```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('19CSE305_LabData_Set3.1.xlsx', sheet_name=1)
```

```python
# fields having '?' is considered as null, remove them
df = df.replace('?', np.nan)

df.isnull().sum()

```

```python
# fill missing values with mean column values
df.fillna(df.mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# fill missing values with random values from column
for col in df.columns:
    if df[col].dtype == 'object':
        df = df.fillna(df[col].value_counts().index[0])
        
df.isnull().sum()
        
```

```python
for column in df.columns:
    data_type = df[column].dtype
    if data_type == 'object':
        data_type = 'Categorical'
    elif data_type == 'int64' or data_type == 'float64':
        data_type = 'Numerical'
    else:
        data_type = 'Ordinal'
        
    print(f"Attribute: {column}, Data Type: {data_type}")



```

```python
# data range for numerical attributes
for column in df.columns:
    data_type = df[column].dtype
    if data_type == 'int64' or data_type == 'float64':
        print(f"Attribute: {column}, Min: {df[column].min()}, Max: {df[column].max()}")
```
