```python
# fields having '?' is considered as null, remove them
df = df.replace('?', np.nan)

df.isnull().sum()

```

```python
# fill missing values with mean column values
df.fillna(df.mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# fill missing values with random values from column
for col in df.columns:
    if df[col].dtype == 'object':
        df = df.fillna(df[col].value_counts().index[0])
        
df.isnull().sum()
        
```

```python
for column in df.columns:
    data_type = df[column].dtype
    if data_type == 'object':
        data_type = 'Categorical'
    elif data_type == 'int64' or data_type == 'float64':
        data_type = 'Numerical'
    else:
        data_type = 'Ordinal'
        
    print(f"Attribute: {column}, Data Type: {data_type}")



```

```python
# fill categorical values with the mode of the column
for col in df.columns:
    if df[col].dtype == 'object':
       if df[col].isnull().any():
           mode_value = df[col].mode().iloc[0]
           df[col] = df[col].fillna(mode_value)

```

```python
# data range for numerical attributes
for column in df.columns:
    data_type = df[column].dtype
    if data_type == 'int64' or data_type == 'float64':
        print(f"Attribute: {column}, Min: {df[column].min()}, Max: {df[column].max()}")
```

```python
# study the presence of missing values in each attribute
for column in df.columns:
    print(f"Attribute: {column}, Missing Values: {df[column].isnull().sum()}")
```

```python
# outliers in numerical attributes
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (data < lower_bound) | (data > upper_bound)


for column in df.columns:
    data_type = df[column].dtype
    if data_type == 'int64' or data_type == 'float64':
        outliers = detect_outliers_iqr(df[column])
        num_outliers = outliers.sum()
        print(f"Attribute: {column}, Outliers: {num_outliers}")
        
# outliers in numerical attributes

        
```

```python
# calculate mean and variance for the numerical attributes

for column in df.columns:
    data_type = df[column].dtype
    if data_type == 'int64' or data_type == 'float64':
        print(f"Attribute: {column}, Mean: {df[column].mean()}, Variance: {df[column].var()}")
```
